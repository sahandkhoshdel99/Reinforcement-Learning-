```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st
from dists import *
```

```python
def mean_confidence_interval(sample_means,confidence_level):
    sample_mean = np.mean(sample_means)
    
    standard_error = np.sqrt(np.var(sample_means))
    z_score = st.norm.ppf(confidence_level/2 + 0.5)
    h = standard_error * z_score
    
    return sample_mean-h,sample_mean, sample_mean+h
```

First Lets look at the biased estimation for std:
- The Maximum Likelihood Estimation (MLE) for standard deviation <b>(Biased</b>):<br><br>
$$\mathit{{\displaystyle \boxed{ s_{n}={\sqrt {{\frac {1}{n}}\sum _{i=1}^{n}\left(x_{i}-{\bar {x}}\right)^{2}}}}}}$$<br><br>
$\textit{Proof of biasness:}$<br><br>
This estimator is derived by taking the square root of the biased estimation for variance so it should probably be biased too, as taking a square root doesn't act as a proper bias correction factor but first let's take a closer look to the MLE (Maximum Likelihood Estimation) for variance which is Biased:<br><br>$$\boxed{\mathit{s_n^2 = \frac {1}{n} \sum_{i=1}^n  \left(x_i - \overline{x} \right)^ 2 }}$$
         <font size="4"><b>How much bias does the MLE of Varince have?</b><br><br><br>
${\displaystyle {\begin{aligned}\boxed{\operatorname {E} \left[\sigma ^{2}-s_{n}^{2}\right]}&=\operatorname {E} \left[{\frac {1}{n}}\sum _{i=1}^{n}(x_{i}-\mu )^{2}-{\frac {1}{n}}\sum _{i=1}^{n}(x_{i}-{\overline {x}})^{2}\right]\\&=\operatorname {E} \left[{\frac {1}{n}}\sum _{i=1}^{n}\left((x_{i}^{2}-2x_{i}\mu +\mu ^{2})-(x_{i}^{2}-2x_{i}{\overline {x}}+{\overline {x}}^{2})\right)\right]\\&=\operatorname {E} \left[{\frac {1}{n}}\sum _{i=1}^{n}\left(\mu ^{2}-{\overline {x}}^{2}+2x_{i}({\overline {x}}-\mu )\right)\right]\\&=\operatorname {E} \left[\mu ^{2}-{\overline {x}}^{2}+{\frac {1}{n}}\sum _{i=1}^{n}2x_{i}({\overline {x}}-\mu )\right]\\&=\operatorname {E} \left[\mu ^{2}-{\overline {x}}^{2}+2({\overline {x}}-\mu ){\overline {x}}\right]\\&=\operatorname {E} \left[\mu ^{2}-2{\overline {x}}\mu +{\overline {x}}^{2}\right]\\&=\operatorname {E} \left[({\overline {x}}-\mu )^{2}\right]\\&=\operatorname {Var} ({\overline {x}})\\&={\boxed{\frac {\sigma ^{2}}{n}}}\end{aligned}}}$<br><br><br><b>Therefore the Expected value of the biased Estimator is:</b><br><br>
$${\displaystyle \boxed{\operatorname {E} \left[s_{n}^{2}\right]}=\sigma ^{2}-{\frac {\sigma ^{2}}{n}}=\boxed{{\frac {n-1}{n}}\sigma ^{2}}}$$
<br><br>

             
[Calculation Source: Wikipedia, Bessel Correction](https://en.wikipedia.org/wiki/Bessel%27s_correction#Proof_of_correctness_%E2%80%93_Alternative_3)

             
<p style = "font-family:Times;">
<font size="3">
<br><br><br>
- The corrrection coefficient factor
$\mathit{(\frac{n}{n-1})}$
is called the Bessel Correction Factor.As a result the Unbiased Variance Estimator is derived from multiplying the Biased estimator ($\mathit{\,s_n^2\,}$) by the "Bessel correction coefficient factor"<br><br>
$$\mathit{\boxed{s^2} = \left(\frac{n}{n-1}\right)\,s_n^2 = \boxed{\frac { \sum_{i=1}^n  \left(x_i - \overline{x} \right)^ 2}{n-1} }}$$<br><br><br>
Primarily, it seems that by taking the square root of the unbiased variance estimator, ($\mathit{s^{2}}$) we can obtain an unbiased estimation for standard deviation but by calculating the expected value of the derived estimator we observe a slight bias that fortunately effects very low sample sizes and tends to zero by increaseing n. So in our case which n = 100, This estimator can be assumed as an unbiased estimator:<br><br>

$$\boxed{\displaystyle s={\sqrt {\frac {\sum _{i=1}^{n}(x_{i}-{\overline {x}})^{2}}{n-1}}}}$$

$$ {\displaystyle \operatorname {Bias} ({\hat {\theta }},\theta )=\operatorname {E} _{x\mid \theta }[\,{\hat {\theta }}-\theta \,]} $$
